# strength surface

template script to generate strength surface of various type
and plot 2d and 3d strength surface from data (and save contours)

```python
from cm_vis.strength.gen import StrengthSurface

mname = 'pmma'
stype = 'VMS'
props = [10]
srange = [-100, 100, 201]
data_dir = f"../data/strength/ss_{mname}_{stype}_props{props}_srange{srange}.npy"

surface = StrengthSurface(stype, props, srange, data_dir)
surface.gen()
```

```python
# %%
from cm_vis.strength.plot import SurfacePlotter

file_name = "../data/strength/ss_pmma_VMS_props[10]_srange[-100, 100, 201]"
data_dir = file_name + ".npy"

plotter = SurfacePlotter(data_dir)
plotter.plot(dim=2, save=False)
```
